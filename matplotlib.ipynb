Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
clr = np.random.rand(50)
sz =1000*np.random.rand(50)
plt.title("random data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y,c=clr,s=sz,alpha = 1)
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(141)
plt.plot(x1, y1, color="orange", marker="*")
plt.subplot(142)
plt.plot(x2, y2, color="orange", marker="*")
plt.subplot(143)
plt.plot(x3, y3, color="orange", marker="*")
plt.subplot(144)
plt.plot(x4, y4, color="orange", marker="*")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
Ans A bar plot is used when you want to show a distribution of data points or perform a comparison of metric
    values across different subgroups of your data. From a bar plot, we can see which groups are highest or most
    common, and how other groups compare against the others.
```

```python
import numpy as np
company = ['Apple', 'Microsoft', 'Google', 'AMD']
profit = [3000, 8000, 1000, 10000]
```

```python
plt.barh(company,profit)
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
```

```python
Ans A box plot is a type of chart that shows the distribution of numerical data and skewness. It displays the 
    minimum, maximum, median, and quartiles (or percentiles) of a data set. A box plot is useful when you want 
    to compare data between multiple groups, show the variability or dispersion of the data, or identify 
    outliers.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
